```python
import numpy as np
import pandas as pd

import tensorflow as tf

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
# plt.style.use("seaborn-darkgrid")
```

```python
x = np.int32(np.linspace(0, 200, 11)) #value of the strike price
y = np.int32(np.linspace(0, 100, 11)) # stock_price
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(x, y)
plt.xlabel("Stock Price")
plt.ylabel("Strike Value")
plt.title("Option Payoff Diagram")
plt.legend()
plt.show()
```

The Call Option Graph

```python
# max(S - E, 0)
# where E = Strike Price
# and S = Stock price 

e = 100
profits = [0 if (num - e) < 0 else (num - e) for num in x]

others = x - e
print(x)
print(others)
print(profits)


fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(x, profits, color="green", label='Profit(S - E)')
plt.xlim(0, 200)
plt.ylim(0, 100)
ax.vlines([100], min(x), max(x), color=["r"],  label='Strike price')
plt.xlabel("Stock Price")
plt.ylabel("Strike Value")
plt.title("Call Option Payoff Diagram")
plt.grid(True)
plt.legend()
plt.show()

```

The Put Option Payoff Diagram

```python
# max(S - E, 0)
# where E = Strike Price
# and S = Stock price 

e = 100
profits = [0 if (e - num) < 0 else (e - num) for num in x]

others = e - x
print(x)
print(others)
print(profits)


fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(x, profits, color="green", label='Profit(E - S)')
plt.xlim(0, 200)
plt.ylim(0, 100)
ax.vlines([100], min(x), max(x), color=["r"],  label='Strike price')
plt.xlabel("Stock Price")
plt.ylabel("Strike Value")
plt.title("Put Option Payoff Diagram")
plt.grid(True)
plt.legend()
plt.show()

```
